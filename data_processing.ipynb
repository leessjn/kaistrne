```python
import cv2
import os
import numpy as np
```

```python
def normalize_image():
    img_normalized = []

    for i in range(1000,10000):
        try:
            filename = f'images/depth_image_{i}.png'
            img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

            img_normalized.append(img)
        except:
            break

    return img_normalized
```

```python
def get_image_able(event, x, y, flag, param):
    image = param
    global able_image_index, img_num

    if event == cv2.EVENT_LBUTTONDOWN:
        width, height = 80, 80
        x1, y1 = max(0, x - width // 2), max(0, y - height // 2)
        x2, y2 = min(image.shape[1], x + width // 2), min(image.shape[0], y + height // 2)

        cropped_image = image[y1:y2, x1:x2]

        file_name = f'cropped_image_{img_num}_{able_image_index}.png'
        save_dir = "./able_image/able_image"
        file_path = os.path.join(save_dir, file_name)
        cv2.imwrite(file_path, cropped_image)
        able_image_index += 1

    return able_image_index
```

```python
def get_image_unable(event, x, y, flag, param):
    image = param
    global unable_image_index, img_num

    if event == cv2.EVENT_LBUTTONDOWN:
        width, height = 80, 80
        x1, y1 = max(0, x - width // 2), max(0, y - height // 2)
        x2, y2 = min(image.shape[1], x + width // 2), min(image.shape[0], y + height // 2)

        cropped_image = image[y1:y2, x1:x2]

        file_name = f'cropped_image_{img_num}_{unable_image_index}.png'
        save_dir = "./unable"
        file_path = os.path.join(save_dir, file_name)
        cv2.imwrite(file_path, cropped_image)
        unable_image_index += 1

    return unable_image_index
```

```python
def get_data():
    img_list = normalize_image()
    print(img_list)

    global able_image_index, unable_image_index, img_num
    img_num = 1000

    for img in img_list:
        able_image_index = 1
        unable_image_index = 1

        cv2.imshow(f'click able image({img_num})', img)
        cv2.setMouseCallback(f'click able image({img_num})', get_image_able, param=img)
        key = cv2.waitKey(0)
        # if key == ord('e'):
        #     break  # Exit the loop if 'e' is pressed
        cv2.destroyAllWindows()
        cv2.imshow(f'click unable image({img_num})', img)
        cv2.setMouseCallback(f'click unable image({img_num})', get_image_unable, param=img)
        key = cv2.waitKey(0)
        if key == ord('e'):
            cv2.destroyAllWindows()
            break  # Exit the loop if 'e' is pressed
        cv2.destroyAllWindows()

        img_num += 1

    return able_image_index - 1, unable_image_index - 1
```

```python
able_img_num, unable_img_num = get_data()
```
